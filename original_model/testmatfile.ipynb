```python
import os
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt

data=loadmat(os.path.join('data/Indian_pines_corrected.mat'))['indian_pines_corrected']
labels=loadmat(os.path.join('data/Indian_pines_gt.mat'))['indian_pines_gt']

print(data.shape)
print(labels.shape)

print('labels')
plt.figure(0)
plt.imshow(labels.reshape(145,145), cmap=plt.cm.gray, interpolation="nearest")

print('bands')
for i in range(1,100,20):
    img=data[:,:,i]
    img1 = img.reshape(145,145)
    plt.figure(i)
    plt.imshow(img1, cmap=plt.cm.gray, interpolation="nearest")
```
